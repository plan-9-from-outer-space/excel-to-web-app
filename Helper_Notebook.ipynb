```python
import pandas as pd
import plotly.express as px
print('Done')
```

## ü§ì Load DataFrame

```python
excel_file = 'Survey_Results.xlsx'
sheet_name = 'DATA'
df = pd.read_excel(excel_file, 
                   sheet_name = sheet_name,
                   header = 3,                          
                   usecols='B:D')
df.head()
```

```python
df.shape
```

```python
total_rows = df.shape[0]
print(f'Total Rows: {total_rows}')
```

## üë©‚Äçüíª Filter/Group DataFrame

### Filter by Departments

```python
departments = df['Department'].unique().tolist()
departments
```

```python
departments_selection = ['Sales', 'Marketing', 'Finance']
departments_selection
```

```python
# Filter = mask
mask = df['Department'].isin(departments_selection)
print(df[mask].shape)
df[mask]
```

### Filter by Age 

```python
junk = (30, 40)
junk2 = [*junk]
junk2
```

```python
age_selection = (30, 40)
mask = df['Age'].between(*age_selection)
df[mask]
```

### Group by Rating 

```python
df.columns
```

```python
df_grouped = df.groupby(by=['Rating']).count()[['Age']]
df_grouped
```

### Rename Column

```python
df_grouped = df_grouped.rename(columns={'Age': 'Votes'})
df_grouped = df_grouped.reset_index()
df_grouped
```

## üìà Plot Data

```python
len(df_grouped)
```

```python
bar_chart = px.bar(df_grouped, 
             x = 'Rating', 
             y = 'Votes',
             text = 'Votes', # print count at top of bars
             color_discrete_sequence = ['#F63366'] * len(df_grouped),
             template = 'plotly_white'
             )
```

```python
bar_chart.show()
```
